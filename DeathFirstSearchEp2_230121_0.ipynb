```python
# n_noeuds = 8   # nb de noeuds
n_noeuds = 37   # nb de noeuds

# liens = [[6, 2],[7, 3],[6, 3],[5, 3],[3, 4],[7, 1],[2, 0],[0, 1],[0, 3],[1, 3],[2, 3],[7, 4],[6, 5]]
liens = [[2, 5], [14, 13], [16, 13], [19, 21], [13, 7], [16, 8], [35, 5], [2, 35], [10, 0], [8, 3], [23, 16], [0, 1], [31, 17], [19, 22], [12, 11], [1, 2], [1, 4], [14, 9], [17, 16], [30, 29], [32, 22], [28, 26], [24, 23], [20, 19], [15, 13], [18, 17], [6, 1], [29, 28], [15, 14], [9, 13], [32, 18], [25, 26], [1, 7], [34, 35], [33, 34], [27, 16], [27, 26], [23, 25], [33, 3], [16, 30], [25, 24], [3, 2], [5, 4], [31, 32], [27, 25], [19, 3], [17, 8], [4, 2], [32, 17], [10, 11], [29, 27], [30, 27], [6, 4], [24, 15], [9, 10], [34, 2], [9, 7], [11, 6], [33, 2], [14, 10], [12, 6], [0, 6], [19, 17], [20, 3], [21, 20], [21, 32], [15, 16], [0, 9], [23, 27], [11, 0], [28, 27], [22, 18], [3, 1], [23, 15], [18, 19], [7, 0], [19, 8], [21, 22], [7, 36], [13, 36], [8, 36]]
n_liens = len(liens) # nb de liens

# portes = [4,5]
portes = [0, 16, 18, 26]
n_portes = len(portes)
```

```python
noeuds = []
for i in range(n_noeuds):
    noeuds.append(i)
for lien in liens:
    for noeud in noeuds:
        if noeud in portes:
            noeuds.remove(noeud)
```

```python
def liensDuNoeud(noeud):
    res = []
    for lien in liens:
        if noeud in lien:
            res.append(lien)
    return res

def noeudsDirects(noeud):
    res = []
    for lien in liensDuNoeud(noeud):
        if noeud == lien[0]:
            res.append(lien[1])
        else:
            res.append(lien[0])
    return res

def portesDirectes(noeud):
    res = []
    for noeud in noeudsDirects(noeud):
        if noeud in portes:
            res.append(noeud)
    return res

def cheminsDuNoeud(noeud):
    res = []
    for lien in liensDuNoeud(noeud):
        chemin = lien if lien[0] == noeud else [noeud, lien[0]]
        res.append(chemin)
    return res

def lienDuChemin(chemin):
    for lien in liens:
        if lien == chemin or lien == chemin[::-1]:
            return lien
        
def porteDuLien(lien):
    for noeud in lien:
        if noeud in portes:
            return noeud

def reductionChemin(chemin):
    res = [chemin[0]]
    for i,noeud in enumerate(chemin[:-1]):
        if noeud != chemin[i+1]:
            res.append(chemin[i+1])
    return res

def cheminsPossibles(noeud, prof):
    res = [[noeud]]
    for i in range(prof):
        chemins = res
        res = []
        for chemin in chemins:
            for c in cheminsDuNoeud(chemin[-1]):
                res.append(chemin+c)
    for i,chemin in enumerate(res):
        chemin.pop(0)
        res[i] = reductionChemin(chemin)
#         print(res[i][:prof])
        for e in res[i][:-1]:
            if e in portes:
                res[i] = res[i][:prof]
    return (res)
```

```python
noeud_depart = 2
profondeur   = 3

liste_ch_portes = []

for c in cheminsPossibles(noeud_depart, profondeur):
    for porte in portes:
        if porte in c:
            # alerte car une porte se trouve sur un chemin !
            # pour le moment on se contente de couper le lien associé à l'alerte
            prof   = c.index(porte)
            chemin = c[prof-1:prof+1]
            lien   = lienDuChemin(chemin)
            porte  = porteDuLien(lien)
            nbCheminsPorte = len(cheminsDuNoeud(porte))
            
            liste_ch_portes.append([c, prof, porte, lien, nbCheminsPorte, cheminsDuNoeud(porte)])
```

```python
import pandas as pd
colonnes = ['chemin', 'profondeur', 'porte', 'lien sensible', 'nb chemins porte', 'chemins de la porte']
df_ch_portes = pd.DataFrame(liste_ch_portes, columns=colonnes)
df = df_ch_portes
```
